```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly as ply
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Load in csv
gm_df = pd.read_csv("../Javier_Sada/cars.csv")
gm_df
```

```python
gm_df['date'] = pd.to_datetime(gm_df['date'])
gm_df = gm_df.resample("M", on = "date").agg({"high": "mean", "volume": "mean"}).reset_index()
gm_df
```

```python
fig = go.Figure([go.Scatter(x=gm_df["date"], y = gm_df["high"])])
fig.update_layout(
    title="Time VS Stock",
    xaxis_title="Year",
    yaxis_title="Stock Value USD",
    legend_title="Legend Title",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="slategray"
    )
)
fig.update_layout(
    title={
        'text': "Time VS Stock",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

fig.show()
```
