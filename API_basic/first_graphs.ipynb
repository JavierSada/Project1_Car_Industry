```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly as ply
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Load in csv
gm_df = pd.read_csv("../Javier_Sada/GM_all.csv")
gm_df['Unnamed: 0'] = pd.to_datetime(gm_df['Unnamed: 0'])
gm_df = gm_df.resample("M", on = "Unnamed: 0").agg({"2. high": "mean", "5. volume": "mean"}).reset_index()
gm_df.head()
```

```python
# Load in csv
nissan_df = pd.read_csv("../Javier_Sada/Nissan_all.csv")
nissan_df['Unnamed: 0'] = pd.to_datetime(gm_df['Unnamed: 0'])
nissan_df = nissan_df.resample("M", on = "Unnamed: 0").agg({"2. high": "mean", "5. volume": "mean"}).reset_index()
nissan_df.head()
```

```python
# Load in csv
ford_df = pd.read_csv("../Javier_Sada/Ford_all.csv")
ford_df['Unnamed: 0'] = pd.to_datetime(gm_df['Unnamed: 0'])
ford_df = ford_df.resample("M", on = "Unnamed: 0").agg({"2. high": "mean", "5. volume": "mean"}).reset_index()
ford_df.head()
```
